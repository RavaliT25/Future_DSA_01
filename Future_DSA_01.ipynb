```python
# Conversion of int type to percentage type in Views column
# Firstly convert int to decimal type and then to percentage type
total_views = social_media['Views'].sum()
print(total_views)
# Convert int to decimal
social_media['Views_Percentage'] = (social_media['Views'] / total_views) * 100
social_media[['Views', 'Views_Percentage']].head()
```

```python
# Convert decimal values to percentage format in 'Views_Percentage' column
social_media['Views_Percentage'] = social_media['Views_Percentage'].apply(
    lambda x: str(x) + '%' if '%' not in str(x) else x
)

# Preview the updated DataFrame (specific columns)
social_media[['Views', 'Views_Percentage']]
```

```python
# Find maximum and minimum Likes
max_likes = social_media['Likes'].max()
min_likes = social_media['Likes'].min()

# Find maximum and minimum Shares
max_shares = social_media['Shares'].max()
min_shares = social_media['Shares'].min()

# Find maximum and minimum Comments
max_comments = social_media['Comments'].max()
min_comments = social_media['Comments'].min()

# Print the results
print(f"Maximum Likes: {max_likes}, Minimum Likes: {min_likes}")
print(f"Maximum Shares: {max_shares}, Minimum Shares: {min_shares}")
print(f"Maximum Comments: {max_comments}, Minimum Comments: {min_comments}")
```

SENTIMENTAL ANALYSIS NLP

```python
import nltk
nltk.download('vader_lexicon')  # For sentiment analysis
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Calculate thresholds for views
view_25th = social_media['Views'].quantile(0.25)
view_75th = social_media['Views'].quantile(0.75)

# Function to classify sentiment based on views
def classify_sentiment(views):
    if views >= view_75th:
        return 'Positive'
    elif views <= view_25th:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the function to classify sentiment
social_media['Sentiment_Based_on_Views'] = social_media['Views'].apply(classify_sentiment)

# Preview the results
print(social_media[['Views', 'Sentiment_Based_on_Views']].head())

```

```python
import matplotlib.pyplot as plt

# Pie chart for sentiment distribution
social_media['Sentiment_Based_on_Views'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(8, 8))
plt.title('Sentiment Distribution Based on Views')
plt.ylabel('')  # Remove ylabel for a cleaner look
plt.show()
```

```python
import matplotlib.pyplot as plt

# Bar chart for average engagement per platform
social_media.groupby('Platform')[['Likes', 'Shares', 'Comments']].mean().plot(kind='bar', figsize=(10, 6), color=['skyblue', 'orange', 'green'])
plt.title('Platform-Wise Average Engagement')
plt.xlabel('Platform')
plt.ylabel('Average Engagement')
plt.legend(loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot for sentiment vs engagement
colors = {'Positive': 'green', 'Neutral': 'blue', 'Negative': 'red'}
plt.scatter(social_media['Sentiment_Based_on_Views'], social_media['Likes'], c=social_media['Sentiment_Based_on_Views'].map(colors))
plt.title('Sentiment vs Likes Engagement')
plt.xlabel('Sentiment')
plt.ylabel('Likes')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram for views
plt.hist(social_media['Views'], bins=20, color='purple', edgecolor='black')
plt.title('Views Distribution Across Posts')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Prepare data for heatmap
heatmap_data = social_media.groupby(['Region', 'Platform'])[['Likes', 'Shares', 'Comments']].mean().unstack().fillna(0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Region-wise Engagement Heatmap')
plt.xlabel('Platform')
plt.ylabel('Region')
plt.show()
```
